```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_adjust_axes(adj_x=1, adj_y=1):
    xVal = np.linspace(-1,1,1001)
    yVal = []

    for i in xVal: 
        yVal.append(adjust_axes(val=i, adj_x=adj_x, adj_y=adj_y))

    yVal_reverse = yVal[:]
    yVal_reverse.reverse()

    fig, ax = plt.subplots()
    ax.plot(xVal, yVal)
    ax.plot(xVal, yVal_reverse)

    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data',0))
    ax.spines['right'].set_visible(False)
#     ax.grid()
```

```python
draw_adjust_axes(0.8, 0.5)
```

```python
def draw_comb(adj_x=1, adj_y=1, center_y=1, peak_y=-1):
    xVal = np.linspace(-1,1,1001)
    yAdj = []
    yVal = []
    
    for i in xVal: 
        yAdj.append(adjust_axes(val=i, adj_x=adj_x, adj_y=adj_y))
    
    for i in yAdj: 
        yVal.append(steer_val(val=i, center_y=center_y, peak_y=peak_y))    

    yVal_reverse = yVal[:]
    yVal_reverse.reverse()

    fig, ax = plt.subplots()
    ax.plot(xVal, yVal)
    ax.plot(xVal, yVal_reverse)

    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data',0))
    ax.spines['right'].set_visible(False)
#     ax.grid()
```

```python
draw_comb(adj_x=0.8, adj_y=0.5, center_y=0.8, peak_y=-0.3)
```
